```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
df = pd.read_csv('/Users/dnandi/Documents/pythoncourseexamples/Automobile (1).csv')
```

```python
sns.catplot(data=df, x = 'fuel_type', y='horsepower', hue='fuel_type');
```

```python
sns.catplot(data=df, x = 'body_style', y='horsepower', hue='fuel_type', kind='point');
```

```python
#cat plot can show various categorical data vs numerical data
sns.catplot(data=df, x = 'body_style', y='horsepower', hue='fuel_type', kind='bar', palette='pastel');
```

```python
sns.pairplot(data=df[['normalized_losses', 'wheel_base', 'curb_weight', 'engine_size', 'price', 'peak_rpm']]);
```

```python
sns.set(style="darkgrid");
sns.pairplot(data=df, vars=['wheel_base', 'curb_weight','engine_size', 'price'], hue='number_of_doors');
```
