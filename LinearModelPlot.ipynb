```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
df = pd.read_csv('/Users/dnandi/Documents/pythoncourseexamples/Automobile (1).csv')
df.head(5)
```

```python
#lmplot - Linear model plot and it show the best fit line along with data points plotted in 2D space
sns.lmplot(data=df, x='curb_weight', y='horsepower');
```

```python
sns.lmplot(data=df, x='curb_weight', y='horsepower', hue='fuel_type', ci=False);
```

```python
sns.lmplot(data=df, x='curb_weight', y='horsepower', col='fuel_type', ci=False);
```
