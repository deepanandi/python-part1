```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
df = pd.read_csv('/Users/dnandi/Documents/pythoncourseexamples/Automobile (1).csv')
df.head(5)
```

```python
sns.stripplot(data=df, x='engine_size');
```

```python
sns.stripplot(data=df, x='body_style', y='engine_size', hue='body_style', jitter=True);
```

```python
plt.figure(figsize=(10,7))
sns.stripplot(data=df, x='number_of_doors', y='price', jitter=True);
```

```python
sns.swarmplot(data=df, x='number_of_doors', y='price');
```

```python
sns.swarmplot(data=df, x='fuel_type', y='price', hue='number_of_doors', dodge=True);
```
