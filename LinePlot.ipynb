```python
#multiple y values for x data
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")
flights.head()
sns.lineplot(data=flights, x = 'month' , y = 'passengers');

```

```python
flights.head(100)
sns.lineplot(data=flights, x = 'month' , y = 'passengers', ci=False, hue='year');
```

```python
fmri = sns.load_dataset("fmri")
fmri.head(10)
```

```python

sns.lineplot(data=fmri, x="timepoint", y ="signal", hue="region", style="region", errorbar=('ci',False), markers = True);
plt.ylabel('Number of Passengers')
plt.legend(bbox_to_anchor=[1,1]);
```
